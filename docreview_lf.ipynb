### CHECK NEIGBORHOOD LABEL

```python
def check_neighborhood(index_tree_current, status_error_low, status_error_up, low_label, up_label):
    doc_classses = ['RC2','BLOKI', 'BLOKIII', 'BLOKIV401', 'BLOKIV414', 'BLOKIV428', 'BLOKV', 'BLOKVI', 'BLOKVII1', 'BLOKVII2', 'BLOKVIII', 'CATATAN', 'KONDEF', 'COVER']
    
    if(len(low_label)==0):
        index_class_up = doc_classses.index(up_label)
        new_class_index = index_class_up-1
        return new_class_index
    elif(len(up_label)==0):
        index_class_low = doc_classses.index(low_label)
        new_class_index = index_class_low+1
        return new_class_index

    if((not status_error_low) and (not status_error_up)):
        index_class_low = doc_classses.index(low_label)
        index_class_up = doc_classses.index(up_label)
        diff_index = index_class_up-index_class_low
        if(diff_index == 2):
            new_class_index = index_class_low+1
            return new_class_index
        elif(up_label=="BLOKI"):
            new_class_index = index_class_low+1
            return new_class_index
        elif(up_label=="COVER"):
            new_class_index = index_class_up-1
            return new_class_index
        elif(up_label=="KONDEF"):
            new_class_index = index_class_up-1
            return new_class_index
        else:
            new_class_index = index_class_low+1
            return new_class_index



    elif(not status_error_low):
        print('low class status : true')
        new_class_index = index_class_low+1
        return new_class_index
    elif(not status_error_up):
        print('up class status : true')
        new_class_index = index_class_up-1
        return new_class_index
```
